```python
import pandas as pd
```

```python
home=pd.read_csv('/content/House Price Prediction Dataset.csv')
```

```python
home
```

```python
x=home[['Area', 'Bathrooms', 'Bedrooms', 'Floors' ]]
y=home['Price']
```

```python
y
```

```python

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
from xgboost import XGBRegressor
```

```python
xgb_model = XGBRegressor(
    n_estimators=100,
    learning_rate=0.03,
    max_depth=6,
    subsample=0.8,
    colsample_bytree=0.8,
)
```

```python
xgb_model.fit(x_train, y_train)
```

```python
y_pred = xgb_model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
print('Mean Squared Error', mean_squared_error(y_test, y_pred))
print('R-2 Score:', r2_score(y_test, y_pred))
```

```python
new_house = [[2000, 3, 2, 2]] #area, no of bedrooms, no of floors,
new_house_scaled = sc.transform(new_house)
```

```python
new_house_scaled
```

```python
print("Estimated House price:",xgb_model.predict(new_house_scaled))
```
