```python
import pandas as pd
```

```python
data = pd.read_csv('cleaned_movies.csv')
```

```python
lower = data['genre'] = data['genre'].str.lower()
data['title'] = data['title'].str.title()
```

```python
genre=input("enter genre(e.g.,action,comedy):").lower()
start_year=int(input("enter start year:"))
end_year=int(input("enter end year:"))
```

```python
filtered = data[
    (data['genre'].str.contains(genre, na=True)) &
    (data['year'] >= start_year) &
     (data['year'] <= end_year)]
```

```python
filtered=filtered.sort_values(by='year',ascending=True)
```

```python
for year in sorted(filtered['year'].unique()):
  print(f"\nYear",{int(year)})
  movies = filtered[filtered['year'] == year]

  for title in movies['title'].head(5):
    print(f"{title}")
```
